```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
import pandas as pd
from MLP import Network
```

```python
digits = fetch_openml('mnist_784')
print(digits.data.shape)
```

```python
# one hot encoding the target
target = pd.get_dummies(digits.target).astype(int)
target = target.to_numpy()
target = np.expand_dims(target, axis=1)
target.shape
```

```python
data = digits.data.to_numpy()
data = data / 255
data = np.expand_dims(data, axis=1)
data.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=42)
```

```python
BATCH_SIZE = 32
EPOCHS = 2
```

```python
network = Network([784, 128, 64, 10],["sigmoid", "sigmoid", "softmax"], loss_func="cross_entropy")
```

```python
for epoch in range(EPOCHS):
    for i in range(0, len(X_train), BATCH_SIZE):
        batch_X = X_train[i:i+BATCH_SIZE]
        batch_y = y_train[i:i+BATCH_SIZE]
        network.fit(batch_X, batch_y)
    print(f'Epoch {epoch+1}/{EPOCHS} Loss: {network.losses[-1]}')

network.visualize_loss()
```
